```python
import asyncio
import aiohttp

async def fetch_data(key, i):
    url = f'http://openAPI.seoul.go.kr:8088/{key}/json/SearchOpenAPIIOValueService/1/5/{i}/'
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            req = await response.json()
            return req.get('SearchOpenAPIIOValueService', {}).get('row')

async def main(serviceId, key):
    tasks = []
    async with aiohttp.ClientSession(trust_env=True) as session:
        for i in tqdm(serviceId):
            try:
                task = asyncio.create_task(fetch_data(key, i))
                tasks.append(task)
            except:
                print('1')
                pass
        results = await asyncio.gather(*tasks)
        return results

# temp = asyncio.run(main(serviceId, key))
```
